```python
le=LabelEncoder()
df['ssc_p'] = le.fit_transform(df['ssc_p'])
df['ssc_b'] = le.fit_transform(df['ssc_b'])
df['gender'] = le.fit_transform(df['gender'])
df['hsc_p'] = le.fit_transform(df['hsc_p'])
df['hsc_b'] = le.fit_transform(df['hsc_b'])
df['hsc_s'] = le.fit_transform(df['hsc_s'])
df['degree_t'] = le.fit_transform(df['degree_t'])
df['workex'] = le.fit_transform(df['workex'])
df['specialisation'] = le.fit_transform(df['specialisation'])
df['status'] = le.fit_transform(df['status'])
df










```

```python
df=df.to_numpy() #converting the dataset into numpy array
n_samples,n_features=df.shape
n_features=n_features-1 #used to count all the input features and neglect the output co;umn (last column)
x=df[:,0:n_features] #it only takes all the input features
y=df[:,n_features]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
print('training set size: {}, testing set size: {}'.format(x_train.shape,x_test.shape))
```

```python
model = XGBClassifier(n_estimators=10000,learning_rate=0.84,booster="gblinear")
model.fit(x_train,y_train)
```

```python
y_predict=model.predict(x_test)
df=pd.DataFrame(y_test) #To show in a table (This has been fixed default)
df['Predicted Price']=y_predict #This is an extra attribute in the table.
df=df.rename(columns={0:'Actual Price'})
df
```

```python
cm=confusion_matrix(y_test, y_predict)
cm
```

```python
print(classification_report(y_test, y_predict))
score=round(accuracy_score(y_test, y_predict),4)*100
print('Accuracy score: {}% '.format(score))
```
